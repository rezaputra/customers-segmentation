```python
import pandas as pd
import numpy as np
from hdbscan import HDBSCAN
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import LabelEncoder, StandardScaler
import umap
```

```python
dataset1 = pd.DataFrame(pd.read_csv('dataset/onlineretail.csv.csv'))
dataset2 = pd.DataFrame(pd.read_csv('dataset/bankchurners.csv.csv'))
dataset3 = pd.DataFrame(pd.read_csv('dataset/dwimedialestari.csv.csv'))

dataset1 = dataset1.drop(columns=['Unnamed: 0', 'id'])
dataset2 = dataset2.drop(columns=['Unnamed: 0', 'id'])
dataset3 = dataset3.drop(columns=['Unnamed: 0', 'id'])
```

```python
clean_dataset1 = dataset1.dropna(how='any',axis=0)
clean_dataset1 = dataset1.drop_duplicates()
categorical_features_dataset1 = pd.DataFrame(clean_dataset1.select_dtypes(include=['object']))
numeric_features_dataset1 = pd.DataFrame(clean_dataset1.select_dtypes(include=['int64', 'float64']))

clean_dataset2 = dataset2.dropna(how='any',axis=0)
clean_dataset2 = dataset2.drop_duplicates()
categorical_features_dataset2 = pd.DataFrame(clean_dataset2.select_dtypes(include=['object']))
numeric_features_dataset2 = pd.DataFrame(clean_dataset2.select_dtypes(include=['int64', 'float64']))

clean_dataset3 = dataset3.dropna(how='any',axis=0)
clean_dataset3 = dataset3.drop_duplicates()
categorical_features_dataset3 = pd.DataFrame(clean_dataset3.select_dtypes(include=['object']))
numeric_features_dataset3 = pd.DataFrame(clean_dataset3.select_dtypes(include=['int64', 'float64']))
```

```python
encoded_categorical_features_dataset1 = categorical_features_dataset1.apply(LabelEncoder().fit_transform)
encoded_dataset1 = pd.concat([encoded_categorical_features_dataset1, numeric_features_dataset1], axis=1) 

encoded_categorical_features_dataset2 = categorical_features_dataset2.apply(LabelEncoder().fit_transform)
encoded_dataset2 = pd.concat([encoded_categorical_features_dataset2, numeric_features_dataset2], axis=1) 

encoded_categorical_features_dataset3 = categorical_features_dataset3.apply(LabelEncoder().fit_transform)
encoded_dataset3 = pd.concat([encoded_categorical_features_dataset3, numeric_features_dataset3], axis=1) 
```

```python
scaler = StandardScaler()
scale_dataset1 = scaler.fit_transform(encoded_dataset1)
scale_dataset2 = scaler.fit_transform(encoded_dataset2)
scale_dataset3 = scaler.fit_transform(encoded_dataset3)
```

```python
scale_dataset1
```

```python
pca = PCA(n_components=2)
pca_dataset1 = pca.fit_transform(scale_dataset1)
pca_dataset2 = pca.fit_transform(scale_dataset2)
pca_dataset3 = pca.fit_transform(scale_dataset3)
```

```python
plt.scatter(pca_dataset1[:, 0], pca_dataset1[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Plot')
```

```python
plt.scatter(pca_dataset2[:, 0], pca_dataset2[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Plot')
```

```python
plt.scatter(pca_dataset3[:, 0], pca_dataset3[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Plot')
```

```python
reducer = umap.UMAP(n_components=2, metric='euclidean')

umap_dataset1 = reducer.fit_transform(scale_dataset1)
umap_dataset2 = reducer.fit_transform(scale_dataset2)
umap_dataset3 = reducer.fit_transform(scale_dataset3)
```

```python
plt.scatter(umap_dataset1[:, 0], umap_dataset1[:,1])
plt.xlabel('UMAP1')
plt.ylabel('UMAP2')
plt.title('UMAP Plot')
```

```python
plt.scatter(umap_dataset2[:, 0], umap_dataset2[:,1])
plt.xlabel('UMAP1')
plt.ylabel('UMAP2')
plt.title('UMAP Plot')
```

```python
plt.scatter(umap_dataset3[:, 0], umap_dataset3[:,1])
plt.xlabel('UMAP1')
plt.ylabel('UMAP2')
plt.title('UMAP Plot')
```

```python
Elbow_M = KElbowVisualizer(KMeans(), k=20, metric='silhouette')
Elbow_M.fit(scale_dataset1)
Elbow_M.show()
```
