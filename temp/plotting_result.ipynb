```python
# dwimedialestari.csv

import matplotlib.pyplot as plt

# Menyiapkan data
clustering_algorithms = ['HDBSCAN', 'HDBSCAN + PCA', 'K-means', 'Affinity Propagation', 'Agglomerative', 'DBSCAN']
silhouette_scores = [0.3310, 0.4337, 0.2417, 0.2249, 0.8108, 0.4288]
execution_times = [0.395, 0.135, 0.022, 2.921, 0.299, 0.189]

# Membuat plot
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.bar(clustering_algorithms, silhouette_scores)
# plt.xlabel('Clustering Algorithms')
plt.ylabel('Silhouette Score')
plt.title('Dataset 3: dwimedialestari.csv')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.bar(clustering_algorithms, execution_times)
# plt.xlabel('Clustering Algorithms')
plt.ylabel('Execution Time (seconds)')
# plt.title('Execution Time for Dataset 1')
plt.xticks(rotation=45)

# Menampilkan plot
plt.tight_layout()
plt.show()
```
